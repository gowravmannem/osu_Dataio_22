```python
def label_day_of_week(x):
    mon = [1,8,15,2,29]
    tue = [2,9,16,23,30]
    wed = [3,10,17,24]
    thu = [4,1,18,25]
    fri = [5,12,19,26]
    sat = [6,13,20,27]
    sun = [7,14,21,28]
    day =""
    if x in mon:
        day = 'Monday'
    elif x in tue:
        day = 'Tuesday'
    elif x in wed:
        day = 'Wednesday'
    elif x in thu:
        day = 'Thursay'
    elif x in fri:
        day = 'Friday'
    elif x in sat:
        day = 'Saturday'
    elif x in sun:
        day = 'Sunday'
    return day
```
