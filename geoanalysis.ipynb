```python
import seaborn as sns
```

```python
summary["distinmiles"]=summary["tripdistance"]/1609
summary["distinmiles"]=summary["distinmiles"].astype(int)
```

```python
sns.scatterplot(data=summary,x="distinmiles",y="avgspeed")
```

```python
summary["filedate"]=summary["filedate"].astype(str)
```

```python
summary["day"]=summary["filedate"].str[6:8].astype(int)
summary["month"]=summary["filedate"].str[4:6]
summary["year"]=summary["filedate"].str[0:4]
del summary["filedate"]
```

```python
summary["day"]=summary["day"].astype(int)
summary["month"]=summary["month"].astype(int)
summary["year"]=summary["year"].astype(int)
```

```python
def label_day_of_week(x):
    mon = [1,8,15,22,29]
    tue = [2,9,16,23,30]
    wed = [3,10,17,24]
    thu = [4,11,18,25]
    fri = [5,12,19,26]
    sat = [6,13,20,27]
    sun = [7,14,21,28]
    day =""
    if x in mon:
        day = 'Monday'
    elif x in tue:
        day = 'Tuesday'
    elif x in wed:
        day = 'Wednesday'
    elif x in thu:
        day = 'Thursay'
    elif x in fri:
        day = 'Friday'
    elif x in sat:
        day = 'Saturday'
    elif x in sun:
        day = 'Sunday'
    return day
```

```python
summary['DayOfWeek'] = summary.apply(lambda x: label_day_of_week(x['day']),axis=1)
```

```python
summary['DayOfWeek']
```

```python
summary['isweekend'] = [True if x in ['Sunday','Saturday'] else False for x in summary['DayOfWeek']]
```

```python
summary[['DayOfWeek','isweekend']]
```
